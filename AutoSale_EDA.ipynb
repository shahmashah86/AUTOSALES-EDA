##6. Dropping irrelevant columns

* This step is certainly needed in every EDA because sometimes there would be many columns that we never use in such cases dropping is the only solution.In this columns such as ORDERNUMBER,PRODUCTCODE,PHONE,CONTACTFIRSTNAME,CONTACTLASTNAME,CITY doesn't make any sense to me so i just dropped for this instance



```python
df = df.drop(['ORDERNUMBER','PRODUCTCODE','PHONE','CONTACTFIRSTNAME','CONTACTLASTNAME','CITY'], axis=1)
df.head(5)  # print the head of the dataframe
```

##7. Renaming the columns

* In this instance, most of the column names are very confusing to read, so I
just tweaked their column names. This is a good approach it improves the readability of the data set.


```python
df.rename(columns={"QUANTITYORDERED":"QTYORDERD",'ORDERLINENUMBER':"OLN","PRODUCTLINE":"PL","DAYS_SINCE_LASTORDER":"DSL"},inplace=True)
print(df)
```

##10.Conditional Filtering

* Conditional filtering filter out data such that it meets certain criteria.



* Below  is a dataframe that has a quantity orderd greater than twenty,price   greater than hundred and status of the product which are shipped are filtered out




```python
C1=(df["QTYORDERD"]>20) & (df["PRICEEACH"]>100) & (df["STATUS"]=="Resolved")
print(df[C1])
```

* Below the dataframe is grouped on the basis of product line.And it also shows that on the basis of product line the data frame can be grouped into seven groups



```python
print(df.groupby("PL").ngroups)
print(df.groupby("PL").groups)
print(df.groupby("PL").groups.keys())
```

* Below a dataframe is filtered out that has quantity ordered greater than fifty and status of the product that are shipped




```python
df1=(df.groupby([df["QTYORDERD"]>50 ,"STATUS"]))
print(df1.groups)
print(df1.get_group((True,"Shipped")))
```

```python
df2=df.groupby(["ORDERDATE","PL","STATUS"]).groups
print(df2)
```

* Below a dataframe is filtered out by grouping based on the resolved planes on 01/03/2020



```python
df2=df.groupby(["ORDERDATE","PL","STATUS"])
print(df2.get_group(("01/03/2020","Planes","Resolved")))
```

##11.Visualization for Data Analysis

* Scatterplot with x axis quantityordered and y axis with sales



```python
import seaborn as sns
sns.set(rc={"figure.figsize":(6, 6)})
sns.scatterplot(data=df,x="QTYORDERD",y="SALES")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={"figure.figsize":(6, 6)})
sns.boxplot(df["QTYORDERD"])
plt.legend(loc="best")
```

```python
q1 = df['QTYORDERD'].quantile(0.25)
q3 = df['QTYORDERD'].quantile(0.75)
print("75th quartile: ",q1)
print("25th quartile: ",q3)
iqr = q3 - q1
print ("IQR: ",iqr)
lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr
lower_limitoutliers = df[df['QTYORDERD'] < lower_limit]
upper_limitoutliers = df[df['QTYORDERD'] > upper_limit]
new_df= df[df['QTYORDERD'] < upper_limit]
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={"figure.figsize":(6, 6)})
sns.boxplot(new_df["QTYORDERD"])
plt.legend(loc="best")
```

```python
sns.set(rc={"figure.figsize":(12,10)})
sns.boxplot(data=new_df,x=df["STATUS"],y=new_df["QTYORDERD"],hue="PL")
plt.legend(loc="best")
print(df.shape)
print(new_df.shape)
```
